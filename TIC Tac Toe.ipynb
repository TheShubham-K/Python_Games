```python
#let's import the library

from easyAI import TwoPlayersGame, AI_Player, Negamax
from easyAI.Player import Human_Player
```

```python
#inherit a class from TwoplayerGame class
class Tic_Tac_Toe(TwoPlayersGame):
    
    #define a constructor or initialiser
    def __init__(self, players):
        #define the player and the player who is going to start the game 
        self.players = players
        self.nplayer = 1
        
        #define the type of the board
        
        self.board = [0] * 9
    
    #define the possible moves
    def possible_moves(self):
        return[x + 1 for x, y in enumerate(self.board) if y == 0]
    
    
    #define the move of a player
    def make_move(self, move):
        self.board[int(move) -1] = self.nplayer
        
    
    #lets boost the AI, define when a player makes amove
    def umake_move(self, move):
        self.board[int(move - 1)] = 0
        
    #lets define a lose condition that an opponents have three in a line
    def condition_for_lose(self):
        possible_combination = [[1,2,3],[4,5,6],[7,8,9],
                               [1,4,7],[2,5,8],[3,6,9],[1,5,9],[3,5,7]]
        
        return any([all([(self.board[z - 1] == self.nopponent)
                   for z in combination]) for combination in possible_combination])
    
    #lets define the a check for the finish of the game
    def is_over(self):
        return (self.possible_moves() == []) or self.condition_for_lose()
    
    #lets define a function which shows the current position of the players in the game 
    def show(self):
        print('\n' + '\n'.join([' '.join([['.','0','x'][self.board[3* j + i]]
                               for i in range(3)]) for j in range(3)]))
        
    #compute the score
    def scoring(self):
        return -100 if self.condition_for_lose() else 0
    
```

```python
#define the main method which define the algorithm and start the game
if __name__ == "__main__":
    algo = Negamax(7)
    Tic_Tac_Toe([Human_Player(), AI_Player(algo)]).play()
```
